```python
import random
import pandas as pd


# Define the possible values for each column
gender_values = ["Male", "Female", "Other"]
age_category_values = ["18-24", "25-34", "35-44", "45-54", "55-64", "65-74"]
area_values = ["Central", "Eastern", "Northern", "North Central", "North Western", "Sabaragamuwa", "Southern", "Uva", "Western"]

# Generate random values for the new columns
df['gender'] = random.choices(gender_values, k=len(df))
df['age_category'] = random.choices(age_category_values, k=len(df))
df['area'] = random.choices(area_values, k=len(df))
```
