```python
import pandas as pd
import re

# Define rules for aspect extraction
rules = {
    'price': [r'expensive|price|cost|affordable|reasonable|cheap|value|money|cost-effective|overpriced|budget-friendly|economical|discounted|worth the price|affordable range|bargain|competitive|wallet-friendly'],
    'quality': [r'good|quality|high-grade|reliable|poor|low-grade|durability|sturdywell-made|Long-lasting|premium|strong|love|work|help|long|use|excellent|top-notch|superior|inconsistent|unreliable|high-end|performance|flawless|satisfaction|super|solid|dependable|satisfactory|exceptional|performance|trust'],
    'delivery cost': [r'shipping|cost|delivery|charges|shipment|ontime|free shipping|transportation charges|courier expenses|express delivery|handling fees|international|overnight|rates|discounts|policies|threshold|shipping tracking|time'],
    'packaging': [r'packaging|package|wrapped|box|bottle|cover|presentation|casing|container|design|aesthetic|secure|eco-friendly|damage-proof|protective|appealing|attractive|recyclable']
} 

# Function to apply rules and extract aspects from text
def extract_aspects(text):
    aspects = []
    for aspect, patterns in rules.items():
        for pattern in patterns:
            if re.search(pattern, text, flags=re.IGNORECASE):
                aspects.append(aspect)
                break
    return aspects

# Apply aspect extraction to the dataset
df['aspects'] = df['preprocess_text1'].apply(extract_aspects)

# Convert aspects to binary columns
aspects = ['price', 'quality', 'delivery cost', 'packaging']
for aspect in aspects:
    df[aspect] = df['aspects'].apply(lambda x: 1 if aspect in x else 0)

```
