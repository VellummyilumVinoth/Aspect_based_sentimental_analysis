```python
import csv
import nltk
import pandas as pd

# Download necessary NLTK resources
nltk.download('vader_lexicon')
nltk.download('punkt')

from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize

# Initialize the sentiment analysis model
sia = SentimentIntensityAnalyzer()

# Define the path to your CSV file
csv_file = 'output.csv'

# Define the aspects
aspects = ['price', 'quality', 'delivery cost', 'packaging']

# Define the sentiment labels
sentiment_labels = ['very good', 'good', 'neutral', 'bad', 'poor']

# Read the CSV file and extract reviews and aspects
data = []

with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        product_name = row['product_name']
        review = row['review']
        aspect_list = eval(row['aspects'])

        aspect_sentiments = {'price': '-', 'quality': '-', 'delivery cost': '-', 'packaging': '-'}

        for aspect in aspect_list:
            aspect_lower = aspect.lower()

            if aspect_lower in aspects:
                sentiment_score = sia.polarity_scores(review)['compound']
                sentiment_label = None

                if sentiment_score >= 0.6:
                    sentiment_label = 'very good'
                elif sentiment_score >= 0.2:
                    sentiment_label = 'good'
                elif sentiment_score > -0.2:
                    sentiment_label = 'neutral'
                elif sentiment_score > -0.6:
                    sentiment_label = 'poor'
                else:
                    sentiment_label = 'bad'

                aspect_sentiments[aspect] = sentiment_label

        valid_sentiments = [sentiment for sentiment in aspect_sentiments.values() if sentiment != '-']
        
        if valid_sentiments:
            sentiment_scores = [sentiment_labels.index(sentiment) for sentiment in valid_sentiments]
            overall_sentiment_score = sum(sentiment_scores) / len(sentiment_scores)
            overall_sentiment_score1 = int(round(overall_sentiment_score))
            overall_sentiment = sentiment_labels[int(round(overall_sentiment_score))]
        else:
            overall_sentiment = '-'
        
        data.append({
            "Product_name": product_name,
            'Review': review,
            'Price': aspect_sentiments['price'],
            'Quality': aspect_sentiments['quality'],
            'Delivery Cost': aspect_sentiments['delivery cost'],
            'Packaging': aspect_sentiments['packaging'],
            'Overall Sentiment': overall_sentiment,
            'Overall Sentiment Score': overall_sentiment_score1
        })

# Create DataFrame from the collected data
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
output_file = '/content/drive/MyDrive/zidan/aspect_sentiments.csv'
df.to_csv(output_file, index=False)

print(f"The aspect sentiments have been saved to {output_file}.")
```
