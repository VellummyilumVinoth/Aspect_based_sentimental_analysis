```python
import pandas as pd
import re

# Define rules for aspect extraction
rules = {
    'price': [r'expensive|price|cost|affordable|reasonable|cheap|value|money|'],
    'quality': [r'good|quality|high-grade|reliable|poor|low-grade|durability|sturdywell-made|Long-lasting|premium|strong'],
    'delivery cost': [r'shipping|cost|delivery|charges|shipment|ontime|free shipping'],
    'packaging': [r'packaging|package|wrapped|box|bottle|cover']
} 

# Function to apply rules and extract aspects from text
def extract_aspects(text):
    aspects = []
    for aspect, patterns in rules.items():
        for pattern in patterns:
            if re.search(pattern, text, flags=re.IGNORECASE):
                aspects.append(aspect)
                break
    return aspects

# Apply aspect extraction to the dataset
df['aspects'] = df['preprocess_text'].apply(extract_aspects)
```
